```python
# --- 0) Install once per environment (or add to requirements.txt) ---
# !pip install pandas geopandas sqlalchemy psycopg2-binary shapely folium

import folium
import pandas as pd
import geopandas as gpd
from sqlalchemy import create_engine

# --- 1) DB connection ---
DB_URL = "postgresql+psycopg2://ais:aispass@localhost:5432/ais"
engine = create_engine(DB_URL)

# --- 2) Load areas & gates as GeoDataFrames (geometry comes from PostGIS) ---
sql_area = """
    SELECT area_id, name, kind, subtype, "group", notes, geom
    FROM public.area
"""
sql_gate = """
    SELECT gate_id, area_id, name, kind, subtype, "group", notes, geom
    FROM public.area_gate
"""

gdf_area = gpd.read_postgis(sql_area, con=engine, geom_col="geom")
gdf_gate = gpd.read_postgis(sql_gate, con=engine, geom_col="geom")

# --- 3) Compute a sensible map center (mean of area centroids) ---
if not gdf_area.empty:
    centroids = gdf_area.to_crs(4326).centroid
    center_lat = centroids.y.mean()
    center_lon = centroids.x.mean()
else:
    center_lat, center_lon = 20, 0  # fallback (Atlantic)

# --- 4) Make the base map ---
m = folium.Map(location=[center_lat, center_lon], zoom_start=3, tiles="OpenStreetMap", control_scale=True)

# Utility to add a GeoJSON layer with styling + popup/tooltip
def add_layer(gdf, name, color):
    if gdf.empty:
        return
    gj = folium.GeoJson(
        gdf.__geo_interface__,
        name=name,
        tooltip=folium.GeoJsonTooltip(
            fields=["name", "kind", "subtype", "group"],
            aliases=["Name", "Kind", "Subtype", "Group"],
            sticky=True
        ),
        popup=folium.GeoJsonPopup(
            fields=["name", "notes"],
            aliases=["Name", "Notes"],
            sticky=False,
            max_width=600,
        ),
        style_function=lambda f, c=color: {
            "color": c,
            "weight": 2,
            "fillColor": c,
            "fillOpacity": 0.15,
        },
        highlight_function=lambda f: {"weight": 3, "fillOpacity": 0.25},
        show=True,
    )
    gj.add_to(m)

# --- 5) Split areas into useful layers and add them ---
areas_ports_core      = gdf_area[(gdf_area["kind"]=="port") & (gdf_area["subtype"]=="core")]
areas_ports_approach  = gdf_area[(gdf_area["kind"]=="port") & (gdf_area["subtype"]=="approach")]
areas_lane_corridor   = gdf_area[(gdf_area["kind"]=="lane") & (gdf_area["subtype"]=="corridor")]
areas_cp_corridor     = gdf_area[(gdf_area["kind"]=="chokepoint") & (gdf_area["subtype"]=="corridor")]

add_layer(areas_ports_core,     "Ports — core",        "#1f77b4")
add_layer(areas_ports_approach, "Ports — approach",    "#2ca02c")
add_layer(areas_lane_corridor,  "Lanes — corridor",    "#ff7f0e")
add_layer(areas_cp_corridor,    "Chokepoints — corridor", "#d62728")

# --- 6) Gates: split by side for easy toggling ---
def add_gates_layer(gdf, name, color):
    if gdf.empty:
        return
    # Thinner outline for gate polygons
    gj = folium.GeoJson(
        gdf.__geo_interface__,
        name=name,
        tooltip=folium.GeoJsonTooltip(
            fields=["name", "kind", "subtype", "group"],
            aliases=["Name", "Kind", "Subtype", "Group"],
            sticky=True
        ),
        popup=folium.GeoJsonPopup(
            fields=["name", "notes"],
            aliases=["Name", "Notes"],
            max_width=600,
        ),
        style_function=lambda f, c=color: {"color": c, "weight": 2, "fillColor": c, "fillOpacity": 0.10},
        highlight_function=lambda f: {"weight": 3, "fillOpacity": 0.20},
        show=True,
    )
    gj.add_to(m)

gates = gdf_gate
add_gates_layer(gates, "Gates — West",  "#ced812")


# --- 7) Layer control + save ---
folium.LayerControl(collapsed=False).add_to(m)
m.save("ais_areas_map.html")
m
```
